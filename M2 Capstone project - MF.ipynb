# 4. Data Analysis

## 4.1 Logistic Regression

```python
# X and y

X = df.drop('Diabetes_binary', axis=1)
y = df['Diabetes_binary']
```

```python
from sklearn.model_selection import train_test_split

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 42)
```

```python
# Scale data
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

```python
from sklearn.multiclass import OneVsRestClassifier
from sklearn.linear_model import LogisticRegressionCV, LogisticRegression

# Create model
# Initializing the Logistic Regression model
model = OneVsRestClassifier(LogisticRegression(solver='saga',max_iter=1000))
```

```python
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay, RocCurveDisplay, PrecisionRecallDisplay
```

```python
# Model parameters

# Penalty Type
penalty = ['l1', 'l2']

# Use logarithmically spaced C values
C = np.logspace(0, 4, 10)

param_grid = {'estimator__C':C,'estimator__penalty':penalty}
```

```python
from sklearn.model_selection import GridSearchCV

# Obtain C value
grid_model = GridSearchCV(model,param_grid=param_grid, verbose =1)
grid_model.fit(scaled_X_train, y_train)
```

```python
grid_model.best_params_
```

```python
# creating a new model with the parameters found using grid search
model_lr1 = LogisticRegression(C = 1.0, solver='saga',max_iter=1000)
model_lr1.fit(scaled_X_train, y_train)
```

```python
model_lr1.coef_
```

```python
# getting the values for x and y for the barplot

x_bar = X_train.columns.to_list()
y_bar = [coeff for item in model_lr1.coef_ for coeff in item]

# Barplot of coefficients

plt.figure(figsize = (13,5))
ax = sns.barplot(x = x_bar , y = y_bar)
ax.tick_params(axis='x', rotation=90);
```

```python
# Confusion matrix
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay, RocCurveDisplay, PrecisionRecallDisplay
from sklearn.metrics import ConfusionMatrixDisplay

y_pred = model_lr1.predict(scaled_X_test)
cm = confusion_matrix(y_test, y_pred)

# Confusion matrix plot
disp = ConfusionMatrixDisplay(cm, display_labels=model_lr1.classes_)
disp.plot()
plt.show()
```

```python
# Classification report
report = classification_report(y_test, y_pred)
print(report)
```

## 4.2 Random Forest

```python
# Separate X and y
X = df.drop('Diabetes_binary', axis=1)
y = df['Diabetes_binary']

# Split X and y into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)
```

```python
# Import Random Forest classifier
from sklearn.ensemble import RandomForestClassifier

# Create instance of model with default values
rfc = RandomForestClassifier(n_estimators=200,random_state=0)

# Fit  model
rfc.fit(X_train, y_train)
```

```python
y_pred = rfc.predict(X_test)

from sklearn.metrics import accuracy_score

print('Model accuracy score with 100 decision-trees : {0:0.4f}'. format(accuracy_score(y_test, y_pred)))
```

```python
print(classification_report(y_test, y_pred))
```

```python
# View feature scores
feature_scores = pd.Series(rfc.feature_importances_, index=X_train.columns).sort_values(ascending=False)
feature_scores
```

```python
top_features = feature_scores.index[:5]

X_train_selected = X_train[top_features]
X_test_selected = X_test[top_features]
```

```python
param_grid = {
    'n_estimators': [200],  # Number of trees in the forest
    'max_depth': [None, 10, 20],  # Maximum depth of the tree
    'min_samples_split': [2, 5, 10],  # Minimum number of samples required to split an internal node
    'min_samples_leaf': [1, 2, 4],    # Minimum number of samples required to be at a leaf node
}
```

```python
# Initialize the Random Forest classifier
rf_classifier = RandomForestClassifier(random_state=42)

# Initialize GridSearchCV
grid_search_rf = GridSearchCV(estimator=rf_classifier, param_grid=param_grid, cv=5, scoring='accuracy', verbose=2)

# Fit GridSearchCV to the training data
grid_search_rf.fit(X_train_selected, y_train)
```

```python
# Print the best parameters found by GridSearchCV
print(f"Best parameters found: {grid_search_rf.best_params_}")

# Print the best score achieved
print(f"Best score achieved: {grid_search_rf.best_score_}")
```

```python
# Evaluate the model on the test set
best_rf = grid_search_rf.best_estimator_
y_pred_rf = best_rf.predict(X_test_selected)

# Print classification report and confusion matrix
print(confusion_matrix(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))
```

```python
most_frequent_class = y_test.mode()[0]
null_accuracy = (y_test == most_frequent_class).mean() * 100
print(f"Null Accuracy: {null_accuracy:.2f}%")
```

## 5.1 Final Model

```python
top_features = feature_scores.index[:5]
top_features
```

```python
top_features = feature_scores.index[:5]

X_train_selected = X_train[top_features]
X_test_selected = X_test[top_features]
```

```python
X_train_selected.shape, X_test_selected.shape
```

```python
# Create instance of model with default values
rfc = RandomForestClassifier(max_depth=10, min_samples_leaf=1, min_samples_split=10, n_estimators=200,random_state=0)

# Fit  model
rfc.fit(X_train_selected, y_train)
```

```python
y_pred = rfc.predict(X_test_selected)

print('Model accuracy score with 200 decision-trees : {0:0.4f}'. format(accuracy_score(y_test, y_pred)))
```

```python
from sklearn.metrics import roc_curve, roc_auc_score

# Predict probabilities for ROC curve
y_prob = rfc.predict_proba(X_test_selected)[:, 1]

# Calculate ROC curve
fpr, tpr, _ = roc_curve(y_test, y_prob)

# Calculate AUC
roc_auc = roc_auc_score(y_test, y_prob)

# Plot ROC curve
plt.figure(figsize=(10, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC Curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.show()

print(f"AUC: {roc_auc:.2f}")
```
